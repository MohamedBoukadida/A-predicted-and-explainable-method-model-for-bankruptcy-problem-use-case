```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from importlib import reload
from tensorflow.keras import layers, callbacks
```

```python
data = pd.read_csv('C:/Users/ASUS/Desktop/data.csv')
data_train = data.sample(frac=0.7, axis=0)
data_test = data.drop(data_train.index)

X_train = data_train.drop('Bankrupt', axis=1)
y_train = data_train['Bankrupt']
X_test = data_test.drop('Bankrupt',axis=1)
y_test = data_test['Bankrupt']

print('Original data shape was:', data.shape)
print('x_train : ',X_train.shape, 'y_train : ',y_train.shape)
print('x_test : ',X_test.shape, 'y_test : ',y_test.shape)
```

```python
model = keras.Sequential([
    layers.Dense(units=6, activation='relu', input_shape=[95]),
    layers.Dense(units=6, activation='relu'),
    # the linear output layer 
    layers.Dense(units=1,activation='sigmoid'),
])

model.compile(optimizer ='adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy']
             )

model.summary()
```

```python

history = model.fit(X_train,
                  y_train,
                  epochs = 250,
                  batch_size=1000,
                  
                  verbose=1,
                  validation_data = (X_test, y_test))
```

```python
history_df = pd.DataFrame(history.history)
# Start the plot at epoch 5
history_df.loc[5:, ['loss', 'val_loss']].plot()
history_df.loc[5:, ['accuracy', 'val_accuracy']].plot()

print(("Best Validation Loss: {:0.4f}" +\
      "\nBest Validation Accuracy: {:0.4f}")\
      .format(history_df['val_loss'].min(), 
              history_df['val_accuracy'].max()))
```
